```python
import folium
from folium.features import DivIcon

# مركز الخريطة (مثلاً NASA Johnson Space Center قرب هيوستن)
nasa_coordinate = [29.559684, -95.083097]

# إنشاء الخريطة
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Loop لكل موقع إطلاق في launch_sites_df
for index, row in launch_sites_df.iterrows():
    coord = [row['Lat'], row['Long']]
    
    # إضافة دائرة صفراء صغيرة
    folium.Circle(
        location=coord,
        radius=1000,  # نصف القطر بالمتر
        color='#000000',
        fill=True,
        fill_color='yellow',
        fill_opacity=0.5
    ).add_child(folium.Popup(row['Launch Site'])).add_to(site_map)
    
    # إضافة Marker مع اسم الموقع
    folium.map.Marker(
        location=coord,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12pt; color:#d35400;"><b>%s</b></div>' % row['Launch Site']
        )
    ).add_to(site_map)

# عرض الخريطة
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )
# 2️⃣ دالة لحساب المسافة بين نقطتين باستخدام صيغة Haversine
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # نصف قطر الأرض بالكيلومتر

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# 3️⃣ افترضنا أننا اخترنا موقع إطلاق من launch_sites_df (مثال: Kennedy Space Center)
launch_site_lat = 28.573255
launch_site_lon = -80.646895

# 4️⃣ اخترنا أقرب نقطة ساحلية باستخدام MousePosition على الخريطة
coastline_lat = 28.56367
coastline_lon = -80.57163

# 5️⃣ حساب المسافة
distance_coastline = calculate_distance(
    launch_site_lat, launch_site_lon, coastline_lat, coastline_lon
)

# 6️⃣ إضافة Marker على نقطة الساحل المختارة مع عرض المسافة
distance_marker = folium.Marker(
    location=[coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html='<div style="font-size: 12pt; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance_coastline)
    )
)
site_map.add_child(distance_marker)
```

```python
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians
import folium

# دالة لحساب المسافة بين نقطتين
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # نصف قطر الأرض بالكيلومتر
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# مثال: موقع الإطلاق
launch_site_lat = 28.573255
launch_site_lon = -80.646895

# مثال: نقطة قريبة (يمكن أن تكون مدينة، سكة حديد، طريق سريع، ساحل)
point_lat = 28.56367
point_lon = -80.57163
point_name = "Closest Coastline"

# حساب المسافة
distance = calculate_distance(launch_site_lat, launch_site_lon, point_lat, point_lon)

# إضافة Marker على النقطة مع عرض المسافة
distance_marker = folium.Marker(
    location=[point_lat, point_lon],
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html='<div style="font-size: 12pt; color:#d35400;"><b>{}: {:.2f} KM</b></div>'.format(point_name, distance)
    )
)
site_map.add_child(distance_marker)

# رسم خط PolyLine بين موقع الإطلاق والنقطة المختارة
line = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [point_lat, point_lon]],
    weight=2,
    color='blue'
)
site_map.add_child(line)

# عرض الخريطة
site_map
```
